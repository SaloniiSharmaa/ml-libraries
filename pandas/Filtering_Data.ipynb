```python
bios = pd.read_csv(r"C:\Users\saloni\Python_Lib\Pandas\Raw_data\bios.csv")
```

```python
bios.head()
```

```python
bios.loc[bios['height_cm'] > 215 , ['name','height_cm']] # to filter players with height more than 215cm.
bios[bios['height_cm'] > 215] # short hand method
bios[(bios['height_cm'] > 215) & (bios['born_country']== 'USA')] # to filter more than one thing
```

```python
# filters baased on string operations
bios[bios['name'].str.contains('Keith')] # str are case sensitive so use caps where the data has caps.
bios[bios['name'].str.contains('Keith', case = False)] # TO TURN OF CASE SENSITIVITY.
bios[bios['name'].str.contains('Keith|patrick', case = False)] # use the regex syntax to filter out multiple strs.
# bios[bios['name'].str.contains('Keith|patrick', case = False, regex = False)] TO TURN OFF REGEX.
```

```python
# another filters baased on string operations
bios[bios['born_country'].isin(['USA','FRA','GRB']) & (bios['name'].str.startswith('Keith'))]
```

```python
# another method for filtering
bios.query('born_country =="USA" and born_city == "Seat" ')
```
