```python
import pandas as pd
```

```python
coffee = pd.read_csv(r"C:\Users\saloni\Python_Lib\Pandas\Raw_data\coffee.csv")
```

```python
print(coffee)
```

```python
coffee.loc[1,'Units Sold']= 10  # to format/change a value in dataframes
```

```python
coffee.at[0, 'Day']
# a very optimized way to get a single value. note that it cannot be used to get mutiple values.
```

```python
coffee.iat[0,1]
# similar to iloc but only gives a single value
```

```python
coffee.sort_values('Units Sold') # to get in ascending order.
coffee.sort_values('Units Sold', ascending = False) # to get in descending order.
coffee.sort_values(['Units Sold', 'Coffee Type']) # to get multiple parameters
coffee.sort_values(['Units Sold', 'Coffee Type'], ascending=[0,1]) # to set the ascending and descending order while getting multiple parameters.
```
